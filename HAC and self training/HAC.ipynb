```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_excel('p3_data.xlsx')
dataset['gender'] = dataset['gender'].map({'M': 1, 'W': 0})
```

```python
data = dataset.iloc[:,:3]
data = np.array(data)
y_labels = dataset.iloc[:,3]
```

```python
def Euclidean_Distance(x, y):
    dist = (np.sum((x-y)**2))**0.5
    return dist
```

```python
def hc(data,linkage,no_of_clusters):  
    r = data.shape[0]
    c = data.shape[0]
    size = (r, c)
    distance_matrix = np.zeros(size)
    for i in range(r):
        for j in range(c):
            distance_matrix[i][j] = Euclidean_Distance(data[i], data[j])     
    distances = distance_matrix
    distances[distances ==0] = np.inf
    clusters = updating_cluster(distances,linkage) 
    iteration_number = distances.shape[0] - no_of_clusters
    clusters_to_plot = clusters[iteration_number]
    arr = np.unique(clusters_to_plot)
    return clusters_to_plot, arr
```

```python
def updating_cluster(data,linkage):
    clusters = {}
    r = -1
    c = -1
    a = []
    for n in range(data.shape[0]):
        a.append(n)
        
    clusters[0] = a.copy()
    for k in range(1, data.shape[0]):
        min_val = np.inf
        
        for i in range(0, data.shape[0]):
            for j in range(0, data.shape[1]):
                if(data[i][j]<=min_val):
                    min_val = data[i][j]
                    r = i
                    c = j           
        if linkage =="Single":
            for i in range(0,data.shape[0]):
                if(i != c):
                    temp = min(data[c][i],data[r][i])
                    data[c][i] = temp
                    data[i][c] = temp
        elif linkage=="Complete":
             for i in range(0,data.shape[0]):
                if(i != c and i!=r):
                    temp = max(data[c][i],data[r][i])
                    data[c][i] = temp
                    data[i][c] = temp
        elif linkage=="Average" :
             for i in range(0,data.shape[0]):
                if(i != c and i!=r):
                    temp = (data[c][i]+data[r][i])/2
                    data[c][i] = temp
                    data[i][c] = temp
        for i in range (0,data.shape[0]):
            data[r][i] = np.inf
            data[i][r] = np.inf
       
        minimum = min(r,c)
        maximum = max(r,c)
        for n in range(len(a)):
            if(a[n]==maximum):
                a[n] = minimum
        clusters[k] = a.copy()
        
    return clusters
```

```python
def two_cluster(clusters, no_of_clusters):
    c1 = []
    c2 = []
    k = list(np.unique(no_of_clusters))
    for i in range(len(clusters)):
        if clusters[i] == k[0]:
            c1.append(i)
        else:
            c2.append(i)
    return c1, c2
```

```python
def two_cluster_accuracy(c1, c2):
    m_c1 = 0
    w_c1 = 0
    m_c2 = 0
    w_c2 = 0
    for i in range(len(c1)):
        if dataset.iloc[i, -1] == 1:
            m_c1 += 1
        else:
            w_c1 += 1        
    for i in range(len(c2)):
        if dataset.iloc[i, -1] == 1:
            m_c2 += 1
        else:
            w_c2 += 1
    c1_p = 0
    c2_p = 0
    if m_c1>w_c1:
        c1_p = 1
    if m_c2>w_c2:
        c2_p = 1
    
    correct = 0
    for i in range(len(c1)):
        if dataset.iloc[i, -1] == c1_p:
            correct += 1
    for j in range(len(c2)):
        if dataset.iloc[i, -1] == c2_p:
            correct += 1
    return (correct/len(dataset))*100.0
    
```

```python
def four_cluster(clusters, no_of_clusters):
    c1 = []
    c2 = [] 
    c3 = []
    c4 = []
    k = list(np.unique(no_of_clusters))
    for i in range(len(clusters)):
        if clusters[i] == k[0]:
            c1.append(i)
        elif clusters[i] == k[1]:
            c2.append(i)
        elif clusters[i] == k[2]:
            c3.append(i)
        else:
            c4.append(i)
    return c1, c2, c3, c4
```
